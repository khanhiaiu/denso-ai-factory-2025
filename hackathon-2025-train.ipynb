```python
import os

def create_project_structure(base_dir="datasets"):
    """
    Tạo cấu trúc thư mục cho dự án Anomaly Detection (RealIAD + DTD)
    """
    # Danh sách các đường dẫn cần tạo (chỉ cần liệt kê thư mục con cuối cùng)
    target_folders = [
        # --- Nhánh DTD ---
        "dtd/images/banded",
        "dtd/images/blotchy",
        # Bạn có thể thêm các loại texture khác ở đây nếu cần (braided, bubbly, etc.)
        
        # --- Nhánh RealIAD / PCB5 ---
        "RealIAD/PCB5/DISthresh/good",
        "RealIAD/PCB5/ground_truth/bad_mask",
        "RealIAD/PCB5/test/bad",
        "RealIAD/PCB5/test/good",
        "RealIAD/PCB5/train/good"
    ]

    print(f"🚀 Đang tạo cấu trúc thư mục tại: {os.path.abspath(base_dir)}\n")

    for folder in target_folders:
        # Ghép đường dẫn gốc với thư mục con
        full_path = os.path.join(base_dir, folder)
        
        # Tạo thư mục (exist_ok=True giúp không báo lỗi nếu thư mục đã có sẵn)
        os.makedirs(full_path, exist_ok=True)
        print(f"  ✅ Đã tạo: {full_path}")

# Hàm kiểm tra và in ra cây thư mục kết quả
def print_tree(startpath):
    print("\n📂 KẾT QUẢ CẤU TRÚC THƯ MỤC:")
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        indent = '│   ' * (level)
        print(f'{indent}├── {os.path.basename(root)}/')

# --- CHẠY ---
if __name__ == "__main__":
    create_project_structure()
    print_tree("datasets")
```

```python
import os
import shutil
import glob
from tqdm.notebook import tqdm

# ================= CẤU HÌNH ĐƯỜNG DẪN =================
# Đường dẫn nguồn (Input)
INPUT_DIR = "/kaggle/input/hackathon-2025"

# Đường dẫn đích (Output - Working directory)
OUTPUT_DIR = "/kaggle/working/denso-ai-factory-2025/datasets/RealIAD/PCB5/train/good"

# ================= THỰC HIỆN SAO CHÉP =================

# 1. Tạo thư mục đích nếu chưa có
os.makedirs(OUTPUT_DIR, exist_ok=True)
print(f"✅ Đã tạo/kiểm tra thư mục đích: {OUTPUT_DIR}")

# 2. Tìm tất cả các file trong thư mục nguồn (đệ quy)
# Tìm tất cả file có đuôi ảnh phổ biến
extensions = ['*.jpg', '*.jpeg', '*.png', '*.bmp', '*.tif', '*.tiff']
files_to_copy = []

print("🔍 Đang quét file...")
for ext in extensions:
    # Tìm kiếm đệ quy (recursive=True) để lấy file trong cả thư mục con
    files_to_copy.extend(glob.glob(os.path.join(INPUT_DIR, '**', ext), recursive=True))

print(f"📂 Tìm thấy {len(files_to_copy)} file ảnh.")

# 3. Copy file
count_success = 0
count_fail = 0

print("🚀 Bắt đầu sao chép...")
for src_path in tqdm(files_to_copy, desc="Copying"):
    try:
        # Lấy tên file gốc
        file_name = os.path.basename(src_path)
        dst_path = os.path.join(OUTPUT_DIR, file_name)
        
        # Copy file
        shutil.copy2(src_path, dst_path)
        count_success += 1
        
    except Exception as e:
        print(f"❌ Lỗi sao chép {file_name}: {e}")
        count_fail += 1

# ================= KẾT QUẢ =================
print("\n" + "="*40)
print(f"✅ HOÀN TẤT!")
print(f"📊 Thành công: {count_success} file")
print(f"❌ Thất bại: {count_fail} file")
print(f"📍 Đã lưu tại: {OUTPUT_DIR}")
print("="*40)

# Kiểm tra nhanh số lượng file trong thư mục đích
print(f"Số lượng file trong {OUTPUT_DIR}: {len(os.listdir(OUTPUT_DIR))}")
```
